```python
import pandas as pd
data = pd.read_pickle(r"D:/Uni/SS20/DataScience1/project/DataScienceProject/data_other.pkl")
data.head()
```

```python
df = data.drop(['Bahnhofsviertel','Berkersheim']) # rm missing values
X = df[list(df)[0:5:]]

display(X.head())
Y = df[list(df)[10::]] # CDUbis V-Partei³
display(Y.head())
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.multioutput import MultiOutputRegressor
xtrain, xtest, ytrain, ytest=train_test_split(X, Y, test_size=0.25)

```

```python
gbr = GradientBoostingRegressor()
model = MultiOutputRegressor(estimator=gbr)
model.fit(xtrain, ytrain)
score = model.score(xtrain, ytrain)
print("Training score:", score)
ypred = model.predict(xtest)
print(mean_squared_error(ytest, ypred))

```

```python
df_ypred = pd.DataFrame(ypred,index=ytest.index,columns=list(ytest))
df_ypred.head()
```

```python
import matplotlib.pyplot as plt
fig, axes = plt.subplots(len(list(ytest)), sharex=True, sharey=False, figsize=(5,4*len(list(ytest))))

for i,stadtteil in enumerate(list(ytest)):
    print(stadtteil)
    df_stadtteil = pd.DataFrame(index=ytest.index)
    df_stadtteil['predicted'] = df_ypred[stadtteil]
    df_stadtteil['ground truth'] = ytest[stadtteil]
    df_stadtteil.plot.bar(title=stadtteil,ax=axes[i])
display(fig)
#fig.savefig('./output/predict_parties_test.png', format='png')
```

```python
def PollyPlot(xtrain, xtest, y_train, y_test, lr,poly_transform):
    width = 12
    height = 10
    plt.figure(figsize=(width, height))
    
    
    #training data 
    #testing data 
    # lr:  linear regression object 
    #poly_transform:  polynomial transformation object 
 
    xmax=max([xtrain.values.max(), xtest.values.max()])

    xmin=min([xtrain.values.min(), xtest.values.min()])

    x=np.arange(xmin, xmax, 0.1)


    plt.plot(xtrain, y_train, 'ro', label='Training Data')
    plt.plot(xtest, y_test, 'go', label='Test Data')
    plt.plot(x, lr.predict(poly_transform.fit_transform(x.reshape(-1, 1))), label='Predicted Function')
    plt.ylim([0, 50])
    plt.ylabel('Erreichte Prozent')
    plt.xlabel("Median Brutto")
    plt.legend()
```
